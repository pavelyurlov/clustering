ecoli dataset

```python
df_ecoli = pd.read_csv('datasets/ecoli/ecoli-Copy1.data', header=None)
df_ecoli
```

breast cancer

```python
df_bc = pd.read_csv('datasets/breast cancer/breast-cancer.data', header=None)
df_bc
```

```python
np.zeros((2, 2))
```

```python
X_bc = np.zeros((len(df_bc), 16))
```

```python
def sort_key(x):
    if x == 'no':
        return -1
    elif x == '?':
        return 0
    elif x == 'yes':
        return 1
    elif x is None:
        return 0
    elif type(x) is str:
        s = x.split('-')[0]
        if s.isnumeric():
            return int(s)
        else:
            return x
    else:
        return x

columns = [1, 3, 4, 5, 6, 7, 9]
for j, col in enumerate(columns):
    values = sorted(list(set(df_bc[col])), key=sort_key)
    value_to_num = {}
    for i, v in enumerate(values):
        value_to_num[v] = i + 1
    print('column #{}: {}'.format(col, value_to_num))
    for i in range(len(X_bc)):
        X_bc[i, j] = value_to_num[df_bc.loc[i][col]]

ohe_28 = pd.get_dummies(df_bc[[2, 8]])
X_bc[:, 7:] = ohe_28.values
print(X_bc)
```

```python
np.savetxt('datasets/breast cancer/X.csv', X_bc, '%i', delimiter=',')
```

```python
y_bc = df_bc.values[:, 0]

classes_bc = sorted(list(set(y_bc)))
class_to_id_bc = {}
for i, cl in enumerate(classes_bc):
    class_to_id_bc[cl] = i + 1
print(class_to_id_bc)

for i in range(len(y_bc)):
    y_bc[i] = class_to_id_bc[y_bc[i]]

np.savetxt('datasets/breast cancer/y.csv', y_bc, fmt='%i', delimiter=',')
```
